<span style="color:red">TODO: Fix the malformed "CÉDULA" values.</span>

```python
def fix_cedula(x):
    if (re.match("\d{2}-\d{3}", x)) is None:
        new_cedula = x.replace('-','')
        return new_cedula[0:2] + '-' + new_cedula[2:]
    else:
        return x

homicides_df['CÉDULA'] = homicides_df['CÉDULA'].apply(fix_cedula)
```

```python
# Cero cedulas malformadas despues de el fix
cedula_malformed = homicides_df.loc[homicides_df["CÉDULA"].apply(lambda x: (re.match("\d{2}-\d{3}", x) is None))]
cedula_malformed.shape
```

```python
email_malformed = homicides_df.loc[homicides_df["CORREO"].apply(lambda x: (re.match("\w*\d{4}@\w*\.(edu.co|com)", x) is None))]
email_malformed
```

```python
def get_random_string(length):
    # choose from all lowercase letter
    letters = string.ascii_lowercase
    result_str = ''.join(random.choice(letters) for i in range(length))
    return result_str

def fix_email(x):
    if (re.match("\w*\d{4}@\w*\.(edu.co|com)", x)) is None:
        return get_random_string(8) + '1234@colombia.edu.co'
    else:
        return x
    
homicides_df['CORREO'] = homicides_df['CORREO'].apply(fix_email)
```

```python
# Cero emails mal formados despues del fix
email_malformed = homicides_df.loc[homicides_df["CORREO"].apply(lambda x: (re.match("\w*\d{4}@\w*\.(edu.co|com)", x) is None))]
email_malformed.shape
```

<span style="color:red">TODO: Create a function to fix the digitation errors for "DIA" column.</span>

```python
homicides_df["DIA"].unique()
```

```python
dias = ['lunes','martes','miercoles','jueves','viernes','sabado','domingo']
def fix_day(x):
    min_distance = 1000000
    final_day = ''
    for i in dias:
        distance = pylev.levenshtein(x, i)
        if distance < min_distance:
            min_distance = distance
            final_day = i
    return final_day
        
homicides_df['DIA'] = homicides_df['DIA'].apply(fix_day)
homicides_df['DIA'].unique()
```
